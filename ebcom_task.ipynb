```python
#change the type of referenceID column to string
df_provider['referenceID'] = df_provider['referenceID'].astype(str)
df_bill['referenceID'] = df_bill['referenceID'].astype(str)
```

```python
#inner join operation on referenceID 
#calculate the difference between amount in both side
int_df = pd.merge(df_bill, df_provider, how ='inner', on =['referenceID'])
int_df['diff'] = int_df['amount_x'] - int_df['amount_y']
int_df
```

```python
#filter by non zero amount in difference value
rslt_df = int_df[int_df['diff'] != 0]
rslt_df
```

```python
#is all transactionID registered or more NaN values existed ?
rslt_df[rslt_df['transactionID'].isna()]
```

```python
#find the dublication occurrence (cross join)  

duplicate = int_df[int_df.duplicated('referenceID',keep=False)]
duplicate
```
